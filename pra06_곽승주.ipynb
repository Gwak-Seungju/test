#### 문제 1
주어진 webpages.json을 열고, 노래 가사와 제목만 추출하는 코드를 작성하세요. 

힌트 
- BeautifulSoup 모듈을 사용하세요
    - BeautifulSoup 중, soup.find 함수를 활용하세요.
    - soup.find 함수의 output 된 type을 확인해보세요.
    - 제목은 'title'에 위치합니다.
    - 가사는 공통적으로 'div' 중, 'style': 'font-size: 22px;word-break:break-all;'인 부분에 위치합니다.
- Json 모듈을 사용하세요
    - 가사의 경우 answer.json 처럼 텍스트에 대한 처리를 진행해야합니다.
    - json 파일 저장할 경우 인코딩 방식을 확인하세요.
    - json 파일을 answer.json 처럼 보기 좋게 제출해주세요.


```python
import json
from bs4 import BeautifulSoup

with open('web_pages.json', 'r', encoding='utf-8') as f:
  json_data = json.load(f)
my_answer = []
for i in range(len(json_data)) :
  address = json_data[i]['html']
  soup = BeautifulSoup(address, 'html.parser')
  a = soup.find('title')
  b = soup.find('div', {'style': 'font-size: 22px;word-break:break-all;'})
  c = {'제목' : a.text.replace('(+)', '').replace(' | 노래 가사', '').strip(), '가사' : b.text}
  my_answer.append(c)
  
file_path = 'my_answer.json'
with open(file_path, 'w', encoding='utf-8') as f :
  json.dump(my_answer, f, indent='\t', ensure_ascii=False)
```
